```python
# Importing the modules
import cv2
import numpy as np
# Import the necessary function for image display in Colab
from google.colab.patches import cv2_imshow

# Reading the image
image = cv2.imread('/content/image1.png')

# Creating the kernel with numpy
kernel2 = np.ones((5, 5), np.float32)/25

# Applying the filter
img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel2)

# showing the image using cv2_imshow instead of cv2.imshow
cv2_imshow(image)  # Display the original image
cv2_imshow(img)   # Display the blurred image

cv2.waitKey()
cv2.destroyAllWindows()

```

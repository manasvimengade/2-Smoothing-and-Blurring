```python
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

# Reading the image
image = cv2.imread('/content/image1.png')

# Applying the filter
medianBlur = cv2.medianBlur(image, 9)

# Showing the image
cv2_imshow(image)
cv2_imshow(medianBlur)

cv2.waitKey()
cv2.destroyAllWindows()
```
